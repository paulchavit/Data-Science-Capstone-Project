```python
num_papers = {}

for year in range(2010,2021):
    year = str(year)
    n_papers = len(df[df.code.str.contains(year)])
    year = int(year)
    num_papers.update({year:n_papers})
    
num_papers 
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline


plt.figure(figsize = (10,7))
plt.bar(num_papers.keys(),num_papers.values(),color = 'red')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylabel("number of presentations".upper(),fontsize=16) 
plt.xlabel("year".upper(), fontsize=16)
```

```python
from wordcloud import WordCloud, STOPWORDS 

#preprocessing for word cloud generation
stopwords = set(STOPWORDS)
words = '' 

for val in df.abstract: 
      
    #convert to string the abstracts
    val = str(val) 
  
    #split text into words and symbols (tokens) 
    tokens = val.split() 
      
    #Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    words += " ".join(tokens)+" "
  
```

```python
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='black', 
                stopwords = stopwords, 
                min_font_size = 10).generate(words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```

```python
#http://www.sei.dost.gov.ph/images/stsd/coecod.pdf
#percentage of presentations from the different CHED COEs in physics
nip = len(df[df.affiliation.str.contains("National Institute of Physics")])/len(df)*100
admu = len(df[df.affiliation.str.contains("Ateneo de Manila University")])/len(df)*100
dlsu = len(df[df.affiliation.str.contains("De La Salle University Manila")])/len(df)*100
usc = len(df[df.affiliation.str.contains("University of San Carlos")])/len(df)*100
iit = len(df[df.affiliation.str.contains("Iligan Institute of Technology")])/len(df)*100

coe = {'NIP':nip, 'AdMU':admu, 'USC':usc, 'DLSU':dlsu, 'IIT':iit}
print(coe)
```
